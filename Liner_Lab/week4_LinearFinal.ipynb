```python
from sympy import *
```

```python
init_printing(use_unicode=True)
A = Matrix([[2,0,0],[1,4,-1],[-2,-4,4]])
A
```

```python
P, D = A.diagonalize()
pprint(P)
pprint(D)
(P**-1)*A*P
```

```python
eigenvectors = A.eigenvects()
eigenvectors
```

```python
sum = len(eigenvectors[0][2]) + len(eigenvectors[1][2])
print("All count is  : ",sum)
```
