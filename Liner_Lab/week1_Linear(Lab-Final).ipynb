```python
import numpy as np
import matplotlib.pyplot as plt
```

# 1.)

```python
def transfrom(A,X):
    return np.dot(A,X) # เป็นการคำนวณผลคูณของเมทริกซ์หรือเวกเตอร์
A = np.array([[1,0],[0,-1]]) 
theta = np.linspace(0, np.pi/3, 100)
xvals = np.cos(theta)*np.sin(3*theta)
yvals = np.sin(theta)*np.sin(3*theta)
X = np.row_stack((xvals, yvals))  #รูปเริ่มต้น
ax = transfrom(A,X)
plt.axis("equal") #plt.axis((xL,xR,yL,yR))
plt.grid(True)
plt.plot(X[0,:],X[1,:],"-")
plt.show()
```

# 2.)

```python
A = np.array([[1,0],[0,0]])
A1 = np.array([[0.5,1],[0,2]])
a = transfrom(A,X)
a2 = transfrom(A1,X)
plt.axis("equal") #plt.axis((xL,xR,yL,yR))
plt.grid(True)
plt.plot(a[0,:],a[1,:],"-")
plt.plot(a2[0,:],a2[1,:],"-")
plt.plot(X[0,:],X[1,:],"-")
plt.show()
```

# 3.)

```python
b1=np.array([[2,3],[4,5]])
b2=np.array([[1,3],[5,5]])
b3=np.array([[4,3],[9,5]])
o= transfrom(b1,X)
o1= transfrom(b2,X)
o2= transfrom(b3,X)
plt.axis("equal") #plt.axis((xL,xR,yL,yR))
plt.grid(True)
plt.plot(o[0,:],o[1,:],"r")
plt.plot(o1[0,:],o1[1,:],"y")
plt.plot(o2[0,:],o2[1,:],"b")
plt.show()
```

```python
B = np.array([[0,-1],[1,0]])
print(B)
p = transfrom(B,X)
p1 = transfrom(np.linalg.matrix_power(B,2),X)
p2 = transfrom(np.linalg.matrix_power(B,3),X)
plt.axis("equal") #plt.axis((xL,xR,yL,yR))
plt.grid(True)
plt.plot(X[0,:],X[1,:],"oy") #ตัวเริ่มต้น
plt.plot(p[0,:],p[1,:],"or")
plt.plot(p1[0,:],p1[1,:],"og")
plt.plot(p2[0,:],p2[1,:],"ob")
plt.show()
```
